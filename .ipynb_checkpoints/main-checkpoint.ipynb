```python
from grid_world_env import GridWorldEnv

import matplotlib.pyplot as plt
import numpy as np
import random
import time
```

```python
env = GridWorldEnv()
```

```python
nb_states = env.nb_states
nb_actions = env.nb_actions
```

```python
nb_states, nb_actions
```

# Initialize Q-Table

```python
qtable = np.zeros((nb_states, nb_actions))
```

# Set The HyperParameters

```python
nb_episodes = 5_000
max_nb_steps = int( 1.5 * nb_states * nb_actions )

gamma = 0.98

lr = 0.7

epsilon = 1.0
epsilon_min = 0.5
epsilon_decay_rate = 0.0001
```

# Implement Q Learning

```python
rewards = []

for episode in range(nb_episodes):
    state = env.reset()
    ep_rewards = []
    
    for step in range(max_nb_steps):
        # Select the action
        if random.uniform(0, 1) < epsilon or not qtable[state, :].any(): # EXPLORATION
            action = random.randrange(0, nb_actions)
        else:                                # EXPLOITATION
            action = np.argmax(qtable[state, :])

        # Send the action to the environment and retrieve the new_state, reward, done
        next_state, reward, done = env.step(action)
        ep_rewards.append(reward)

        # Update Q value
        qtable[state, action] = (1 - lr) * qtable[state, action] + lr * (reward + gamma * np.max(qtable[next_state, :]))
        
        if done:
            break

        state = next_state

    # Add Reward For Each Episode
    rewards.append(sum(ep_rewards))

    # Decay The Epsilon
    epsilon = max(epsilon_min, epsilon - epsilon_decay_rate)

    # Render 500 Times
    if episode % (nb_episodes//500) == 0:
        env.render()

    if episode % (nb_episodes//10) == 0:
        print(f"{100*episode/nb_episodes}% | Episode: {episode} | Epsilon: {epsilon:.3f} | Mean Reward: {np.mean(rewards[-nb_episodes//10:])}")
        
env.close()
```

# Visualizing the Rewards Variation

```python
plt.plot( [i for i in range(len(rewards))], rewards, label="Reward" )
plt.title("Rewards Evolution While Training")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
plt.legend()
plt.show()
```

# Visualize The Performance Of The Agent

```python
test_episodes = 1
test_steps = np.sum( env.grid == 0 )
```

```python
sum_rewards = 0

for episode in range(test_episodes):
    state = env.reset()

    for step in range(test_steps):
        env.render()
        time.sleep(0.3)

        action = np.argmax(qtable[state, :])

        next_state, reward, done = env.step(action)
        sum_rewards += reward

        if done:
            break

        state = next_state

print(sum_rewards)
env.close()
```
